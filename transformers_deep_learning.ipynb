```python
import ast
all_val = []
for item in df['genres']:
    obj = ast.literal_eval(item)
    if obj != []:
      df_cast = pd.DataFrame(obj)
      all_val.extend(df_cast['name'].tolist())

len(all_val)
u , c = np.unique(np.array(all_val),return_counts=True)
```

```python
nested_cols = ['belongs_to_collection','genres','spoken_languages','production_companies',
                'production_countries','Keywords','cast','crew']

for col in nested_cols:
  print(col, '\n', df[col][10], '\n\n')
```

```python
# change columns type from string to dict/ list and add {} if there is null value - we will change it back to null after
import ast
for column in nested_cols:
        df[column] = df[column].apply(lambda x: {} if pd.isna(x) else ast.literal_eval(x))
```

```python
# open nested columns and return null values when needed
df['belongs_to_collection'] = df['belongs_to_collection'].apply(lambda d: None if d == {} else d['name'])

df['genres'] = df['genres'].apply(lambda l: [d['name'] for d in l] if len(l)>0 else None)

df['spoken_languages'] = df['spoken_languages'].apply(lambda l: [d['iso_639_1'] for d in l] if len(l)>0 else None)

df['production_companies'] = df['production_companies'].apply(lambda l: [(d['name'], d['origin_country']) for d in l] if len(l)>0 else None)

df['production_countries'] = df['production_countries'].apply(lambda l: [d['iso_3166_1'] for d in l] if len(l)>0 else None)

df['Keywords'] = df['Keywords'].apply(lambda l: [d['name'] for d in l] if len(l)>0 else None)

# the below two columns are transformed to a list of dictionaries (still nested), but only with relevant information.
# if you want to use this column, you can further open the nested columns using the explanation here: https://stackoverflow.com/questions/38231591/split-explode-a-column-of-dictionaries-into-separate-columns-with-pandas
df['cast'] = df['cast'].apply(lambda l: [{'character': d['character'], 'gender': d['gender'], 'name': d['name']} for d in l] if len(l)>0 else None)

df['crew'] = df['crew'].apply(lambda l: [{'department': d['department'], 'gender': d['gender'], 'job': d['job'], 'name': d['name']} for d in l] if len(l)>0 else None)

```

```python
import datasets
import random
import pandas as pd
from IPython.display import display, HTML

def show_random_elements(dataset, num_examples=10):
    assert num_examples <= len(dataset), "Can't pick more elements than there are in the dataset."
    picks = []
    for _ in range(num_examples):
        pick = random.randint(0, len(dataset)-1)
        while pick in picks:
            pick = random.randint(0, len(dataset)-1)
        picks.append(pick)

    df = pd.DataFrame(dataset[picks])
    for column, typ in dataset.features.items():
        if isinstance(typ, datasets.ClassLabel):
            df[column] = df[column].transform(lambda i: typ.names[i])
    display(HTML(df.to_html()))
```

```python
import numpy as np
from sklearn.metrics import hamming_loss, precision_recall_fscore_support, f1_score

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def compute_metrics(eval_pred):
    predictions, labels = eval_pred

    # prediction = model output = logits, trained with BCEWithLogitsLoss
    # BCEWithLogitsLoss: logits -> sigmoid -> BCELoss
    # therefore, to get probabilities, with apply sigmoid
    # multi-label -> therefore using >= 0.5 threshold for each class
    # (if this was a single-label classification, we'd simply use np.argmax)
    predictions = (sigmoid(predictions) >= 0.5).astype(labels.dtype)

    # convert labels back from float to int (see section Encoding the data above)
    # predictions = predictions.astype(np.int32)
    # labels = labels.astype(predictions.dtype)

    # precision, recall, f1 - per class
    p, r, f1, _ = precision_recall_fscore_support(labels, predictions)
    f1_macro = f1_score(labels, predictions, average='macro')

    f1_metrics = {
        f'{l}_f1': _f1
        for l, _f1 in zip(id2label, f1)
    }
    f1_metrics['f1_macro'] = f1_macro

    # hamming loss
    hamming_loss_metrics = {
        'hamming_loss': hamming_loss(labels, predictions),
    }

    # exact match
    exact_match_metrics = {
        'exact_match': (predictions == labels).all(axis=1).sum() / len(labels)
    }

    metrics = {
        **f1_metrics,
        **hamming_loss_metrics,
        **exact_match_metrics,
    }

    return metrics
```

```python
from transformers import AutoTokenizer
##bert-base-cased
##distilbert-base-uncased"
model_checkpoint = "distilbert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_checkpoint, use_fast=True)
```

```python
tokenizer("Hello, this one sentence!", "And this sentence goes with it.")
```

```python
def preprocess_function(examples):
    return tokenizer(examples["overview"], truncation=True)
```

```python
preprocess_function(dataset[:5])
```

```python
encoded_dataset = dataset.map(preprocess_function, batched=True)
```

```python
new_features = encoded_dataset.features.copy()
new_features['labels'] = datasets.Sequence(datasets.Value('float32'))

encoded_dataset = encoded_dataset.cast(new_features)
```

```python
encoded_dataset.features
```

```python
train_test_dataset = encoded_dataset.train_test_split(0.2)
test_dataset = train_test_dataset['test']

test_dataset
```

```python
# split into train/dev sets
train_dev_datasets = train_test_dataset['train'].train_test_split(0.2)

train_dev_datasets
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

num_labels = len(dataset[0]['labels'])

model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint,
                                                           num_labels=num_labels,
                                                           id2label=id2label,
                                                           label2id=label2id,
                                                           problem_type="multi_label_classification")
```

```python
model_name = model_checkpoint.split("/")[-1]

args = TrainingArguments(
    f"{model_name}-finetuned",
    evaluation_strategy = "steps",
    save_strategy = "steps",
    eval_steps=100,
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=15,
    weight_decay=0.01,
    load_best_model_at_end=True,
    metric_for_best_model='f1_macro',
    report_to='none',
)
```

```python
args
```

```python
trainer = Trainer(
    model,
    args,
    # we use only a subset of data for this to work fast for the example
    # you should use your own full dataset
    train_dataset=train_dev_datasets['train'].shuffle().select(range(1000)),
    eval_dataset=train_dev_datasets['test'].shuffle().select(range(500)),
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
trainer.evaluate(test_dataset.shuffle(seed=6).select(range(100)))
```
